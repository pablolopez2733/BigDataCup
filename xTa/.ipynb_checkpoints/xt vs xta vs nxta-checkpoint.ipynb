```python
import os
import sys
import warnings
warnings.filterwarnings("ignore")
module_path = '../modules/'
if module_path not in sys.path:
    sys.path.append(module_path)
import xTa as xTa
import xT as xT
import nxT as nxT
import pandas as pd
#Load data
data=pd.read_csv("data_pre_xg.csv")
data=data[data['league'] == 'scouting']
#xTa
res = xTa.return_xTa_values(data)
data=pd.concat([data,res],axis=1)
#xT
xTModel = xT.ExpectedThreat(l=16, w=8)
xTModel.fit(data)
data['xT']=xTModel.predict(data)
#nxTa
res = nxT.return_nxTa_values(data)
data=pd.concat([data,res],axis=1)
#Compare
ma=xT.get_move_actions(data)
print(ma[['xT','xTa','nxTa']].corr())
#ma.to_csv('comparing_methods.csv',index=False)
```

```python
a=pd.DataFrame(ma.groupby('Player')['nxTa'].mean()).sort_values(by='nxTa',ascending=False)
```

```python
b=pd.DataFrame(ma.groupby('Player')['xTa'].mean()).sort_values(by='xTa',ascending=False)
```

```python
c=pd.DataFrame(ma.groupby('Player')['xT'].mean()).sort_values(by='xT',ascending=False)
```

```python
pd.concat([a,b,c],axis=1)
```

```python
pd.DataFrame(ma.groupby('Player')['xT'].sum()).sort_values(by='xT',ascending=False)
```

```python
import pandas as pd
import datetime
import warnings
warnings.filterwarnings('ignore')
data=pd.read_csv(r"C:\Users\3791150\Documents\BigDataCup-main\analysis.csv")
```

```python
start =datetime.datetime.now()

data['possession'] = 0
for i in range(0,len(data)-1):
    if data['Period'][i] == data['Period'][i+1]: 
        if data['Team'][i+1] == data['Team'][i]:
            data['possession'][i+1] = data['possession'][i]
        else:
            data['possession'][i+1] = data['possession'][i]+1
    else:
        data['possession'][i+1] = 0
end =datetime.datetime.now()

print(end-start)
```
